```python
# Importing packages for pipeline creation, reproduction, pre trained tokenization, progress bar creation
from transformers import pipeline, set_seed
from datasets import load_dataset, load_from_disk, load_metric
import pandas as pd
import matplotlib.pyplot as plot
from transformers import AutoModelForSeq2SeqLM, AutoTokenizer

import nltk
from nltk.tokenize import sent_tokenize

from tqdm import tqdm
import torch

nltk.download("punkt")
```

```python
# Checking for programming model req for parallel GPU computation
device = "cuda" if torch.cuda.is_available() else "cpu"
device
```

```python
# Loading Googles Pegasus model's varient trained on CNN/Daily mail dataset and its architecture, pretrained weights
# Moved the models location to device for computation
model = "google/pegasus-cnn_dailymail"
tokenize = AutoTokenizer.from_pretrained(model)
model_pegasus = AutoModelForSeq2SeqLM.from_pretrained(model).to(device)
```

```python
# Download and unzip data
!wget https://github.com/entbappy/Branching-tutorial/raw/master/summarizer-data.zip
!unzip summarizer-data.zip
```

```python
# Loading data from disk
dataset_samsum = load_from_disk("samsum_dataset")
dataset_samsum
```

```python
# Splitting the data based on train, test and validation. Checking features, dialogue and summary
split_lengths = [len(dataset_samsum[length]) for length in dataset_samsum]
print(f"Split Lengths: {split_lengths}")
print(f"Features: {dataset_samsum['train'].column_names}")
print("\nDialogue:")

print(dataset_samsum['test'][1]['dialogue'])

print("\nSummary:")

print(dataset_samsum['test'][1]['summary'])
```

```python
# Converting the data into feature by tokenizing it to fed ML model
def convert_example_to_features(example_batch):
  input_embeddings = tokenize(example_batch['dialogue'], max_length = 1024, truncation = True)
  with tokenize.as_target_tokenizer():
    target_embeddings = tokenize(example_batch['summary'], max_length = 128, truncation = True)

  return {

      'input_ids' : input_embeddings['input_ids'],
      'attention_mask' : input_embeddings['attention_mask'],
      'labels' : target_embeddings['input_ids']

  }
```

```python
# Applying the function to samsum dataset
dataset_samsum_f = dataset_samsum.map(convert_example_to_features, batched = True)
```

```python
# Checking the data
dataset_samsum_f['train']
```

```python
# Training of model
from transformers import DataCollatorForSeq2Seq
seqtoseq_data_collator = DataCollatorForSeq2Seq(tokenize, model = model_pegasus)
```

```python
# Defining and customizing various training params and hyperparams
from transformers import TrainingArguments, Trainer
trainer_args = TrainingArguments(output_dir='pegasus-samsum', evaluation_strategy='steps',
                                per_device_train_batch_size=1, per_device_eval_batch_size=1,
                                 gradient_accumulation_steps=16, weight_decay=0.01, num_train_epochs=1,
                                 warmup_steps=500, logging_steps=10, eval_steps=500, save_steps=1e6)
```

```python
# Training of arguments
trainer = Trainer(model=model_pegasus, args=trainer_args, data_collator=seqtoseq_data_collator,
                  train_dataset=dataset_samsum_f['train'], eval_dataset=dataset_samsum_f['validation'],
                  tokenizer=tokenize)
```

```python
# Training
trainer.train()
```

```python
# Evaluation
# Generate batch sizes from data
def generate_batch_sized_chunks(list_of_elements, batch_size):
  for i in range(0, len(list_of_elements), batch_size):
    yield list_of_elements[i : i+batch_size]

# Calculate the metric based on evaluation data. Iterate over no of batches to tokenize input and generate summaries
def calculate_metric_on_test_ds(dataset, metric, model, tokenizer, batch_size=16,
                                column_text='article', column_summary='highlights', device=device):
  article_batches = list(generate_batch_sized_chunks(dataset[column_text], batch_size))
  target_batches = list(generate_batch_sized_chunks(dataset[column_summary], batch_size))

  for article_batch, target_batch in tqdm(zip(article_batches, target_batches), total=len(article_batches)):
    inputs = tokenizer(article_batch, max_length=1024, truncation=True, return_tensors='pt', padding='max_length')

    summaries = model.generate(input_ids=inputs['input_ids'].to(device),
                               attention_mask=inputs['attention_mask'].to(device),
                               length_penalty=0.8, num_beams=8, max_length=128)

    decode_summaries = [tokenizer.decode(s, skip_special_tokens=True,
                                         clean_up_tokenization_spaces=True) for s in summaries]


    decode_summaries = [d.replace("", " ") for d in decode_summaries]

    metric.add_batch(predictions=decode_summaries, references=target_batch)

  score = metric.compute()
  return score
```

```python
# ROUGE metric for checking quality of text summarizer sys
rouge_names = ['rouge1', 'rouge2', 'rougeL', 'rougeLsum']
rouge_metric = load_metric('rouge')
```

```python
# Calculate rouge_score
score = calculate_metric_on_test_ds(dataset_samsum['test'], rouge_metric, trainer.model, tokenize,
                                    batch_size=2, column_text='dialogue', column_summary='summary')

rouge_dict = dict((rn, score[rn].mid.fmeasure) for rn in rouge_names)

pd.DataFrame(rouge_dict, index=[f'pegasus'])
```
